## Partícula en un campo magnético

Un electrón con una capacidad de trabajo (energía cinética) de $82090×10^{−19}$ J orbita de manera perpendicular a un campo magnético de 4250 G. ¿Cuál es el radio de la órbita? ¿Y su frecuencia y periodo angular? Resuelve el mismo ejercicio para un antineutrón.

Cuando una carga atraviesa perpendicularmente un campo magnético experimenta una fuerza que le produce una desviación en la dirección, es decir, se curva la trayetoria de esa carga.

La fuerza experimentada será perpendicular al vector dirección que lleva esa partícula.



La ley de Lorentz establece que una partícula __cargada__ que circula a una velocidad $\vec {v},$ por un punto en el que existe una intensidad de campo magnético $\vec{B}$, sufrirá la acción de una fuerza $\vec{F}$ denominada __fuerza de Lorentz__ cuyo valor es proporcional al valor de __q__,$\vec{B}$ y $\vec{v}$. Se obtiene por medio de la siguiente expresión; 
$$\vec{F}= q⋅(\vec{v}×\vec{B}).$$

De la ecuación anterior podemos extrar las siguiente conclusiones:

- Su módulo es $F=|q|⋅v⋅B⋅sin α$ , donde $α$ es el ángulo formado entre $\vec{v}$ y $\vec{B}$

- Su dirección es perpendicular al plano que forman $\vec{v}$ y $\vec{B}$

- Su sentrido será el de $\vec{v}x\vec{B}$ si $q$ es positivo y al contrario si es negativo.



En el caso que se nos pide, $\vec{v}$ y $\vec{B}$ son perpendiculares $sin 90$ = 1 por lo que

$F = |q|·v·B· sin 90 = |q|·v·B$

donde aplicamos que $F = \frac {m v^2}{r}$ (ley de Newton)

$q_e v_e B= \frac {m_e * v_e^2}{r}$

### ¿Cuál es el radio de la órbita?

Teniendo los valores de

B = Campo magnético 4250G = 0.425 Tesla

$v$ = Velocidad de la partícula en $\frac{d(m)}{t(s)}$

$r$ = radio en metros

$q$ = carga en Culombios $1.6·10^{-19}$

$m$ = masa en Kg = $9.1·10^{-31}kg$ 

E = $82090×10^{−19}$ J 


La capacidad de trabajo es la energía cinética

$ E_c = \frac {1}{2}m v^2$

Despejando en esa ecuación

$ 2E_C = m v^2 $    y     $ v = \frac {\sqrt{2 E_c}} {\sqrt{{m}} } $

ahora sustituyendo en la de Lorenz tenemos que:

$ \frac {\sqrt{2 E_c} B q_e} {\sqrt{m_e}} = \frac {2 E_c}{r}$

y despejando el radio, que es lo que queremos calcular, obtenemos

$r = \frac {\sqrt{2 E_c m}} {B q} $

$r = \frac {\sqrt{2 82090·10^{−19}·9.1·10^{-31}}} {0.425·1.6·10^{-19}} = 0.0017975 = 1.179·10^{-3}m$

### ¿Y su frecuencia y periodo angular?

$f$ = Frecuencia, número de vueltas por segundo en Hertzios (Hz)

T = Periodo , Tiempo que se tarda en dar una vuelta completa

Dado que el periodo es la inversa de la frecuencia

$T = \frac {1}{f}$

Sabiendo que el perido angular es


$T = \frac {2 \pi r } {v} = \frac {2 \pi m} {q B} $

Calculamos

$T = \frac {2 \pi . 9.1·10^{-31}} {1.6·10^{-19} · 0.425} = 8.404·10^{-11}s$

$f = \frac {1}{T} = \frac {1}{8.404·10^{-11}} = 1.19·10^{10}Hz$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib inline
```

```python
q = 1.6*10**-19 # carga
m = 9.1*10**-31 # masa
```

```python
E = np.array([1,0,0]) # campo eléctrico
B = np.array([0.3,0,0.3]) # campo magnético
vinit = np.array([0,2,0]) # velocidad inicial de la partícula
```

```python
tinit = 0 # tiempo inicial
tend = 1 # tiempo final
dt = 1e-3 # intervalo de tiempo
niter = int((tend-tinit)/dt) # número de pasos en los que se calcula la ecuación de movimiento
t = np.linspace(tinit,tend,niter) # tiempos en los que se calcula la ecuación de movmimiento
```

```python
v = np.zeros((len(t),3)) # velocidades en cada instante de tiempo
r = np.zeros((len(t),3)) # posiciones en cada instante de tiempo

v[0] = vinit
r[0] = np.array([1,2,3])
```

```python
for i in range(1,niter):
    a = (q)*(E + np.cross(v[i-1],B)) # F = ma = q(E + v x B)
    v[i] = v[i-1] + a*t[i]
    r[i] = r[i-1] + v[i]*t[i]
```

```python
plt.figure(figsize=(8,8))
plt.plot(r[:,0],r[:,1])
plt.xlabel(r"$r_x$ [Unidades arbitrarias]",size=20)
plt.ylabel(r"$r_y$ [Unidades arbitrarias]",size=20)
```

```python
plt.figure(figsize=(8,8))
plt.plot(r[:,0],r[:,2])
plt.xlabel(r"$r_x$ [Unidades arbitrarias]",size=20)
plt.ylabel(r"$r_z$ [Unidades arbitrarias]",size=20)
```
